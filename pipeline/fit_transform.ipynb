```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.base import BaseEstimator, TransformerMixin#build class
from sklearn.utils.validation import check_is_fitted

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X.sum(axis=1)
        if self.add_bedrooms_per_room:
            bedrooms_per_room = (X.sum(axis=1))**2
            return np.c_[X, rooms_per_household,bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household]

class ColumnsSelector(BaseEstimator, TransformerMixin):
    # initializer 
    def __init__(self, columns):
        # save the features list internally in the class
        self.columns = columns
        
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        # return the dataframe with the specified features
        return X[self.columns]



num_pipeline = Pipeline([
        ('columns selector',ColumnsSelector(['ret','turnover','pe'])),
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='median')),
        ('attribs_adder', CombinedAttributesAdder()),
        #('std_scaler', StandardScaler()),
    ])

pd.DataFrame(num_pipeline.fit_transform(data), columns=list(data.columns[2:])+['sum','sum^2'],index = list(data.index.values))
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

all_=['name','ret','turnover','pe']
mark_=all_[0]
var_=[i for i in all_ if i not in mark_]


class group_mean(BaseEstimator, TransformerMixin):
    def __init__(self,by,variable): 
        self.by=by
        self.variable=variable
       
    def fit(self, X, y=None):
        self.map = X.groupby(self.by)[self.variable].mean()
        return self  
    def transform(self, X, y=None):
        X[self.variable] = X[self.variable].fillna(value = X[[self.by]].merge(self.map, on=self.by, how='left'))
        return X
num_pipeline = Pipeline([
        ('columns selector',ColumnsSelector(all_)),
        ('imputer', group_mean(mark_,var_)),
])

pd.DataFrame(num_pipeline.fit_transform(data), columns=all_,index = list(data.index.values))
```
